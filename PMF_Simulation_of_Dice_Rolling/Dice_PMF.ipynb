```python
import random
import matplotlib.pyplot as plt
import numpy as np
```

```python
class DiceExperiment:
    def __init__(self, pmf):
        self.pmf = pmf

    def build_roulette_wheel(self):
        total_prob = sum(self.pmf.values())
        cumulative_distribution = []
        cdf = 0
        for number, prob in self.pmf.items():
            cdf += prob / total_prob
            cumulative_distribution.append((cdf, number))
        return cumulative_distribution

    def roulette_wheel_selection(self, cumulative_distribution):
        r = random.random()
        for cdf, number in cumulative_distribution:
            if r <= cdf:
                return number

    def plot_pie_chart(self):
        labels = self.pmf.keys()
        sizes = self.pmf.values()
        colors = plt.cm.Paired(np.arange(len(labels)))

        plt.figure(figsize=(8, 8))
        plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%',
                startangle=140, wedgeprops={'edgecolor': 'black'})
        plt.title('Roulette Wheel (Probability Pie Chart)')
        plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
        plt.show()
```

```python
# Define the PMF for the dice
pmf = {1: 2/6, 2: 1/12, 3: 1/12, 4: 1/12, 5: 1/12, 6: 2/6}

# Initialize the dice experiment
experiment = DiceExperiment(pmf)
cumulative_distribution = experiment.build_roulette_wheel()
```

```python
# Plot the pie chart
experiment.plot_pie_chart()
```

```python
# Simulate the dice rolls
num_simulations = 10000
results = []

for _ in range(num_simulations):
    result = experiment.roulette_wheel_selection(cumulative_distribution)
    results.append(result)
```

```python
# Count the occurrences of each number
counts = {number: results.count(number) for number in pmf.keys()}

# Calculate the empirical probabilities
empirical_probabilities = {number: count / num_simulations for number, count in counts.items()}

# Print the results
print("Expected PMF:", pmf)
print("Empirical PMF:", empirical_probabilities)
```

```python
# Plot the results
fig, ax = plt.subplots()
numbers = list(pmf.keys())
expected_probs = list(pmf.values())
empirical_probs = [empirical_probabilities[number] for number in numbers]

bar_width = 0.40
index = np.arange(len(numbers))

ax.bar(index, expected_probs, bar_width, label='Expected PMF')
ax.bar(index + bar_width, empirical_probs, bar_width, label='Empirical PMF')

ax.set_xlabel('Dice Face')
ax.set_ylabel('Probability')
ax.set_title('Expected vs Empirical PMF')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(numbers)
ax.legend()

plt.show()
```
